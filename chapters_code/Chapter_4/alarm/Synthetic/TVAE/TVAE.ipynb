```python
import pandas as pd

df_original = pd.read_csv("/alarm/alarm.csv")

df_100 = pd.read_csv("/alarm/Samples/stratified_100.csv")
df_100.head(3)
```

```python
from sdv.metadata import SingleTableMetadata

metadata = SingleTableMetadata()
metadata.detect_from_dataframe(data=df_original)
print('Auto detected data:\n')
metadata
```

```python
metadata.save_to_json('/alarm/Synthetic/metadata.json')
```

```python
from sdv.single_table import TVAESynthesizer
model_100 = TVAESynthesizer(metadata)
model_100.fit(df_100)
sports_100_syn = model_100.sample(num_rows=2000, output_file_path="/alarm/Synthetic/TVAE/alarm_100_syn.csv")
```

```python
df_500 = pd.read_csv("/alarm/Samples/stratified_500.csv")
model_500 = TVAESynthesizer(metadata)
model_500.fit(df_500)
sports_500_syn = model_500.sample(num_rows=10000, output_file_path="/alarm/Synthetic/TVAE/alarm_500_syn.csv")
```
